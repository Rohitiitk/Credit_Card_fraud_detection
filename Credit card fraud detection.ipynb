```python
import os
import sys
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from imblearn.over_sampling import SMOTE
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import classification_report
```

```python
df = pd.read_csv('/kaggle/input/creditcardfraud/creditcard.csv')
df.head(5)
     

# df.info()
     

df0 = df[df['Class']==0]
df1 = df[df['Class']==1]
# Sample from the fraud class to balance the dataset using SMOTE
smote = SMOTE(random_state=2)
X_resampled, y_resampled = smote.fit_resample(df.drop('Class', axis=1), df['Class'])
     

# print(df.shape)
# print(df0.shape)
# print(df1.shape)

     

df.groupby('Class').mean()
     

df1_sample = df1.sample(n=284315, replace=True)
df1_sample
     

df_final=pd.concat([df0,df1_sample],axis=0)
# df_final
     

df_final.groupby('Class').mean()
     

X = df_final.drop(['Class'],axis=1)
y = df_final['Class']
     

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2, stratify=y, random_state=2)
     

model = LogisticRegression(random_state=2)
model.fit(X_train,y_train)

# # Print classification report
# y_pred = model.predict(X_test)
# print(classification_report(y_test, y_pred))

print(model.score(X_test,y_test))

# Print classification report
y_pred = model.predict(X_test)
print(classification_report(y_test, y_pred))
     

import pickle
with open("Credit_Card_Fraud_Detection_Model.pickle","wb") as f:
  pickle.dump(model,f)
```
